2]a) Plotting the effect that the matrix M has on a vector, v. I selected a simple linear vector for this example but the components can be futzed to see other (maybe more interesting) vectors. <br>

$$M= \begin{bmatrix}
  2 & 1 \\\\ 
  1 & 2 
 \end{bmatrix}$$

```python
#=======
' 2]a) '
#=======

#_________________
# width of domain

w = 5

#___________________________
# number of points in array

n = 2 * w

#__________________
# array assignment

sg = np.linspace ( -w , w , n )

#______________________________
# using meshgrid to generate a
# square grid of values

x , y = np.meshgrid ( sg , sg )

#_______________________________
# the x and y components of M*v

Mx = (2 * x + y)
My = (x + 2 * y)

#_________________________
# visualization macguffin

plt.quiver ( x , y  , x , y , color = 'b' , label = 'v' )
plt.quiver ( x  , y  , Mx , My , color='r' , label = r'M$\cdot$v' )

plt.title('The effect of M on v')

plt.xlabel ( 'x' )
plt.ylabel ( 'y' )

plt.xlim ( -w - 2 , w + 2 )
plt.ylim ( -w - 2 , w + 2 )

plt.legend()
plt.grid ( ls = '--' , alpha = 0.6)
```

Now, we look to plot the eigenspace of M. In order to do this we need the eigenvalues and eigenvectors. This is done by solving <br>

$\left|M-\lambda*I\right|=0$ <br>

Which reduces in this case to, <br>

$(2-\lambda)^2-1=\lambda^2-4\lambda+3=(\lambda-1)(\lambda-3)=0$ <br>

Which shows us that $\lambda_1=1,\lambda_2=3$. To go about generating the eigenvectors, we evaluate our calculated eigenvalues and dedcue which vectors adhere to the relation. <br>

$Mv\big|_{\lambda_1}\equiv M_1v_1=\begin{bmatrix} 1 & 1 \\\\ 1 & 1 \end{bmatrix}\begin{bmatrix}v_{11} \\\\ v_{12} \end{bmatrix}=0$

And vice versa for $\lambda_2$. In the above case, it can be shown via row reduction that there is an implied free parameter in this matrix due to the linear dependence in the columns of the matrix. Due to this we can see that $v_1=\begin{bmatrix}1 \\\\ -1\end{bmatrix}$ and through similar analysis that $v_2=\begin{bmatrix}1 \\\\ 1\end{bmatrix}$. Now, for the coding...

```python
#=======
' 2]a) '
#=======

M = [ [ 2 , 1 ] , [ 1 , 2 ] ]

ival , ivec = np.linalg.eig ( M )


s = '|| IGNval | IGNvec (xcomp) | IGNvec (ycomp) ||'

print ( '=' * len ( s ) )
print ( s )
print ( '||' , '-' * ( len ( s ) - 6 ) , '||' )

for i , val in enumerate ( ival ) :

    print ( '|| %6.1f | %14.5f | %14.5f ||' % ( val , ivec [ i ] [ 0 ] , ivec [ i ] [ 1 ] ) )

print( '=' * len ( s ) )

print ( 'Notice the vectors are normalized... 1/sqrt(2) = %.5f' % (1/np.sqrt(2)))
```

```python
#_________________________
# visualization macguffin

plt.quiver ( x , y , x , y , color = 'b' , label = 'v' )
plt.quiver ( x , y , Mx , My , color='r' , label = r'M$\cdot$v')

plt.quiver ( ivec[0][0] , ivec[0][1], color = 'g' , label = 'IGNvecs of M' )
plt.quiver ( ivec[1][0] , ivec[1][1], color = 'g' )

plt.xlabel ( 'x' )
plt.ylabel ( 'y' )

plt.xlim ( -w - 2 , w + 2 )
plt.ylim ( -w - 2 , w + 2 )

plt.legend()
plt.grid ( ls = '--' , alpha = 0.6)
```
